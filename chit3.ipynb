```python
q1=df.health.quantile(0.25)
q3=df.health.quantile(0.75)
q1,q3
```

```python
IQR=q3-q1
IQR
```

```python
lower_limit=q1-1.5*IQR
upper_limit=q3-1.5*IQR
lower_limit,upper_limit
```

```python
df['zscore'] = (df.age-df.age.mean())/df.age.std()
df
```
